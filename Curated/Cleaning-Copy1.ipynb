```python
def widths(att,texto):
    dfs['attribute'].explode().value_counts()
    width=dfs[(dfs['question_type'].astype(str)=='freetext')&(dfs['attribute'].astype(str)==f'{att}')]
    widths=width['value'].explode().value_counts().reset_index()
    widths.rename(columns={'index':'curation'},inplace=True)

    rounding(widths, 'curation',f'{texto}')  
    width_list=widths[widths['rounding'].astype(str)!='[]']['curation'].to_list()
    width_curated=curated[curated['attribute'].astype(str)==f'{att}']
    return width_curated[width_curated['value'].isin(width_list)]  

def second(att, texto):
    w=widths(f'{att}',f'{texto}')
    rounds(w,'value')
    w['value']=w['value'].apply(lambda x: re.sub(r'\s\s',' ',str(x))).apply(lambda x: re.sub(r'in in','in',str(x)))
    rounding(w, 'value',f'{texto}') 
    print(len(w))
    fixed=w[w['rounding'].astype(str)=='[]']
    unfixed=w[w['rounding'].astype(str)!='[]']

    fixed[f'Q:{att}']=fixed['value']
    matchfixed=fixed[['external_id',f'Q:{att}']]
    unfixed[f'Q:{att}']=''
    matchunfixed=unfixed[['external_id',f'Q:{att}']]
    
    return matchfixed,matchunfixed
```
